```python
import numpy as np
import matplotlib.pyplot as plt


def u_exact(x):
    return x - np.sinh(x) / np.sinh(1)


def w_i(i):
    return 2 * (-1)**(i+1) / (i * np.pi * (i**2 * np.pi**2 + 1))


def u_h(x, n):
    u = np.zeros_like(x)
    for i in range(1, n+1):
        u += w_i(i) * np.sin(i * np.pi * x)
    return u


x = np.linspace(0, 1, 1000)


u_e = u_exact(x)


n_val = [2, 4, 8]
u_ap = {n: u_h(x, n) for n in n_val}

# Plotando as solues
plt.figure(figsize=(10, 6))
plt.plot(x, u_e, label='Soluo Exata', linewidth=2)
for n in n_val:
    plt.plot(x, u_ap[n], label=f'Aprox. n={n}', linestyle='--')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Comparao das Solues Aproximadas com a Soluo Exata')
plt.legend()
plt.grid(True)
plt.show()
```
